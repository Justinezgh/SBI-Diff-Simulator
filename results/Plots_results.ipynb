```python
import arviz as az
import matplotlib.pyplot as plt
az.style.use("arviz-darkgrid")
import numpy as np 
import pandas as pd
```

```python
# two moons results data  
two_moons_nlp_without_score = pd.read_table("/content/2moons_neg_log_prob_without_score.tsv")
two_moons_nlp_with_score = pd.read_table("/content/2moons_neg_log_prob_with_score.tsv")

two_moons_nlp_without_score = two_moons_nlp_without_score[['step','neg_log_prob']]
two_moons_mean_nlp_without_score = np.array(two_moons_nlp_without_score.groupby(by=['step']).mean()['neg_log_prob'])
two_moons_std_nlp_without_score = np.array(two_moons_nlp_without_score.groupby(by=['step']).std()['neg_log_prob'])

two_moons_nlp_with_score = two_moons_nlp_with_score[['step','neg_log_prob']]
two_moons_mean_nlp_with_score = np.array(two_moons_nlp_with_score.groupby(by=['step']).mean()['neg_log_prob'])
two_moons_std_nlp_with_score = np.array(two_moons_nlp_with_score.groupby(by=['step']).std()['neg_log_prob'])
```

```python
# two moons
x1 = [20,50,100,200,500,1000]
plt.figure(figsize=(6,5),dpi=100)
plt.plot(x1, two_moons_mean_nlp_with_score, c = 'indianred', marker = 'o', label = 'simulations + 1e-5*score')
plt.fill_between(x1,two_moons_mean_nlp_with_score - two_moons_std_nlp_with_score, two_moons_mean_nlp_with_score + two_moons_std_nlp_with_score,  color = 'indianred', alpha = 0.5)
plt.plot(x1, two_moons_mean_nlp_without_score, c = 'C1', marker = 'o', label = 'simulations')
plt.fill_between(x1,two_moons_mean_nlp_without_score - two_moons_std_nlp_without_score, two_moons_mean_nlp_without_score + two_moons_std_nlp_without_score,  color = 'C1', alpha = 0.5)

plt.ylabel('NLPT')
plt.xlabel('Number of training samples')
plt.xscale('log')
plt.legend()
plt.savefig('twomoons_inference.pdf')
```

```python
import jax
from sbids.tasks import lotka_volterra 
from sbids.tasks.utils import get_samples_and_scores
from numpyro.handlers import condition
```

```python
@jax.jit
def get_batch(key, batch_size=1):
    model = lotka_volterra
    (log_probs, samples), scores = get_samples_and_scores(model, key, batch_size=batch_size)
    return samples['theta'], samples['y']
```

```python
truth, observation = get_batch(jax.random.PRNGKey(4))
```

```python
reference_samples=np.load('/content/reference_sample_seed4.npy')
sample_score=np.load('/content/50_lv_with_score_seed17.npy')
sample_without_score=np.load('/content/50_lv_without_score_seed17.npy')

sample_score20=np.load('/content/20_lv_with_score_seed17.npy')
sample_without_score20=np.load('/content/20_lv_without_score_seed17.npy')
```

```python
parameters = [r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$']
c = ChainConsumer()

c.add_chain(np.array(reference_samples), parameters=parameters, name="Reference")
c.add_chain(np.array(sample_without_score), parameters=parameters, name="Simulations")
c.add_chain(np.array(sample_score), parameters=parameters, name="Simulations + 1e-7score")
c.configure(colors=["#455A64", "#FF6F00", "#B71C1C"], shade=True, shade_alpha=0.2, bar_shade=True,
             tick_font_size=8, label_font_size=25, max_ticks=8)

fig = c.plotter.plot(figsize=[10,10], truth=np.array(truth).squeeze(), 
                     extents=[[t - 5*np.std(reference_samples[:,i]), 
                               t + 5*np.std(reference_samples[:,i])] for i,t in enumerate(np.array(truth).squeeze())])
plt.savefig('50contourplot_lotkavolterra.pdf')
```

## Bijector plots

```python
from sbids.bijectors.bijectors import MixtureAffineSigmoidBijector
```

```python
p = np.array([[[0.1,0.3,0.1,0.4,0.1]]])
a = np.array([20,6,50,100,100]).reshape(-1,1,5)
b = np.array([0.1,0.1,0.5,0.8,2]).reshape(-1,1,5)
c = np.array([0.1,0.1,0.01,0.001,0.01]).reshape(-1,1,5) 
bij=MixtureAffineSigmoidBijector(a,b,c,p)
```

```python
x = np.linspace(0,1,1000)
y = jax.vmap(lambda x: bij.forward(x.reshape([1,1])))(x).reshape(-1,)
plt.figure(figsize=(6,5),dpi=100)
plt.plot(x,y, color = 'crimson', label='mixture transformation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.savefig('smooth_transformation.pdf')
```

```python
x = np.linspace(0,1,1000)
y = jax.vmap(lambda x: bij.forward(x.reshape([1,1])))(x).reshape(-1,)
z = jax.vmap(jax.grad(lambda x: bij.forward(x.reshape([1,1])).reshape(-1,).squeeze()))(x)
plt.figure(figsize=(6,5),dpi=100)
plt.plot(x,z, color = 'lightcoral', label='gradient')
plt.xlabel('x')
plt.ylabel('df/dx')
plt.legend()
plt.savefig('smooth_transformation.pdf')
```
