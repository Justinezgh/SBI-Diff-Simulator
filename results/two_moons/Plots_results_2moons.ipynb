```python
import arviz as az
import matplotlib.pyplot as plt
az.style.use("arviz-darkgrid")
import numpy as np 
import pandas as pd
```

```python
c2st_without_score = pd.read_table("/content/c2st_metric_without_score.tsv")
nlp_without_score = pd.read_table("/content/neg_log_prob_without_score.tsv")

c2st_with_score = pd.read_table("/content/c2st_metric_with_score.tsv")
nlp_with_score = pd.read_table("/content/neg_log_prob_with_score.tsv")
```

```python
c2st_without_score = c2st_without_score[['step','c2st_metric']]
mean_c2st_without_score = np.array(c2st_without_score.groupby(by=['step']).mean()['c2st_metric'])
std_c2st_without_score = np.array(c2st_without_score.groupby(by=['step']).std()['c2st_metric'])

nlp_without_score = nlp_without_score[['step','neg_log_prob']]
mean_nlp_without_score = np.array(nlp_without_score.groupby(by=['step']).mean()['neg_log_prob'])
std_nlp_without_score = np.array(nlp_without_score.groupby(by=['step']).std()['neg_log_prob'])

c2st_with_score = c2st_with_score[['step','c2st_metric']]
mean_c2st_with_score = np.array(c2st_with_score.groupby(by=['step']).mean()['c2st_metric'])
std_c2st_with_score = np.array(c2st_with_score.groupby(by=['step']).std()['c2st_metric'])

nlp_with_score = nlp_with_score[['step','neg_log_prob']]
mean_nlp_with_score = np.array(nlp_with_score.groupby(by=['step']).mean()['neg_log_prob'])
std_nlp_with_score = np.array(nlp_with_score.groupby(by=['step']).std()['neg_log_prob'])
```

```python
x = [20,50,100,200,500,1000]
plt.plot(x, mean_c2st_with_score, c = 'C0', marker = 'o', label = 'simulations + 1e-5*score')
plt.fill_between(x,mean_c2st_with_score - std_c2st_with_score, mean_c2st_with_score + std_c2st_with_score,  color = 'C0', alpha = 0.5)
plt.plot(x, mean_c2st_without_score, c = 'C1', marker = 'o', label = 'simulations ')
plt.fill_between(x,mean_c2st_without_score - std_c2st_without_score, mean_c2st_without_score + std_c2st_without_score,  color = 'C1', alpha = 0.5)
plt.ylabel('c2st')
plt.xlabel('Number of training samples')
plt.legend()
```

```python
x = [20,50,100,200,500,1000]
plt.plot(x, mean_nlp_with_score, c = 'C0', marker = 'o', label = 'simulations + 1e-5*score')
plt.fill_between(x,mean_nlp_with_score - std_nlp_with_score, mean_nlp_with_score + std_nlp_with_score,  color = 'C0', alpha = 0.5)
plt.plot(x, mean_nlp_without_score, c = 'C1', marker = 'o', label = 'simulations')
plt.fill_between(x,mean_nlp_without_score - std_nlp_without_score, mean_nlp_without_score + std_nlp_without_score,  color = 'C1', alpha = 0.5)
plt.ylabel('NLL')
plt.xlabel('Number of training samples')
plt.legend()
```
