```python
im1 = jnp.load('/content/affine_nf_prob.npy')
sample1 = jnp.load('/content/affine_nf_sample.npy')
score1 = jnp.load('/content/affine_nf_score.npy')

plt.clf()
x = jnp.stack(jnp.meshgrid(jnp.linspace(0.,1.,128),
                           jnp.linspace(0.,1.,128)),-1)

plt.figure(figsize=(5,5),dpi=100)
plt.contourf(x[...,0],x[...,1],jnp.exp(im1), 100,cmap='Oranges')
quiver(sample1[...,0], sample1[...,1], score1[:,0] ,score1[:,1],width=0.002);
plt.axis('off')
plt.title('Affine NF trained with score')
plt.savefig('affineNF.png', transparent=True, bbox_inches='tight',
        pad_inches = 0)
```
